```python
import json

characters = json.load(open("raw_characters.json"))
```

```python
valid_characters = {}
world_data_num = {}
for world in characters:
    total_data_num = 0
    world_name = world.replace('_', ' ').title()
    valid_characters[world_name] = {}
    for character in characters[world]:
        character_name = character.replace('_', ' ').title()
        data_num = characters[world][character]
        total_data_num += data_num

        if data_num >= 200 or world_name == "Zootopia" or world_name == "Animal Crossing" or world_name == "Spongebob Squarepants (Series)":
            valid_characters[world_name][character_name] = data_num
    if len(valid_characters[world_name]) == 0:
        del valid_characters[world_name]
    else:
        world_characters = list(valid_characters[world_name].items())
        world_characters.sort(key=lambda x: x[1], reverse=True)
        valid_characters[world_name] = dict(world_characters)
        world_data_num[world_name] = total_data_num

world_data_num = list(world_data_num.items())
world_data_num.sort(key=lambda x: x[1], reverse=True)
world_data_num = world_data_num[:]
print(dict(world_data_num).keys())



```
